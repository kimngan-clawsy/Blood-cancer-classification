```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

# Confusion matrix data
cm = np.array([[697, 9, 0],
               [7, 2661, 0],
               [0, 0, 281]])

labels = ['Healthy', 'AML', 'ALL']

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_facecolor('lightgray')

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(ax=ax, cmap='Greys', colorbar=True)

# Make cell text larger
for text in disp.text_.flat:
    text.set_fontsize(20)


colorbar_ax = fig.axes[-1]  
colorbar_ax.tick_params(labelsize=20)  

# Labels and title
ax.set_title("Average Confusion matrix acroos 10 folds", fontsize=20)
ax.set_xlabel("Predicted label", fontsize=20)
ax.set_ylabel("True label", fontsize=20)
ax.tick_params(axis='both', labelsize=20)

plt.tight_layout()
plt.show()
```
