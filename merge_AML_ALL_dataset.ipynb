# Merge ALL 281 samples + AML 3374 samples

```python
import pandas as pd

# Load the CSV files
df1 = pd.read_csv('C:/Class/HK6/IPrj/Datasets/Leukemia_GSE28497.csv')
df2 = pd.read_csv('C:/Class/HK6/IPrj/Datasets/merged_3374_data.csv')
```

```python
df2
```

```python
df1 = df1.drop(columns = 'samples')
```

```python
df1 = df1.rename(columns={'type': 'target'})
```

```python
df1.isnull().sum().sum()
```

```python
df2.isnull().sum().sum()
```

```python
df1['target'] = df1['target'].replace({"B-CELL_ALL" : 2.0,
                                                   "B-CELL_ALL_ETV6-RUNX1" : 2.0,
                                                   "B-CELL_ALL_HYPERDIP" : 2.0,
                                                   "B-CELL_ALL_T-ALL": 2.0,
                                                   "B-CELL_ALL_TCF3-PBX1" : 2.0,
                                                   "B-CELL_ALL_HYPO" : 2.0,
                                                   "B-CELL_ALL_MLL": 2.0,
                                                   })

df1['target'] = df1['target'].astype("int64")
```

```python
df1
```

```python
df2
```

```python
df_AML_ALL = pd.concat([df1, df2], axis=0, ignore_index=True)
df_AML_ALL
```

```python
common_features = df1.columns.intersection(df2.columns)

# Display common features
print("Common features:", common_features)
```

```python
diff_1 = set(df1.columns) - set(df2.columns)
print("Columns in df1 but not in df2:")
for col in diff_1:
    print(col)
```

```python
df1_common = df1[common_features]
df2_common = df2[common_features]

# Concatenate the datasets row-wise
merged_AML_ALL = pd.concat([df1_common, df2_common], axis=0, ignore_index=True)

# Check shape of merged data
print("Merged dataset shape:", merged_AML_ALL.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = merged_AML_ALL.drop('target', axis=1)
y = merged_AML_ALL['target']

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,1], y=X_pca[:,0], hue=y, alpha=0.8)
plt.title("PCA Scatter Plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Separate target and features
y = merged_AML_ALL.iloc[:, 0]
X = merged_AML_ALL.iloc[:, 1:]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA with 3 components
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
explained_var = pca.explained_variance_ratio_

# Prepare DataFrame for plotting
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
df_pca['target'] = y

# 3D Scatter Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

targets = df_pca['target'].unique()
colors = sns.color_palette('Set2', n_colors=len(targets))

for target, color in zip(targets, colors):
    subset = df_pca[df_pca['target'] == target]
    ax.scatter(subset['PC1'], subset['PC2'], subset['PC3'], label=target, s=60, alpha=0.8, color=color)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D PCA Scatter Plot')
ax.legend()
plt.tight_layout()
plt.show()
```
